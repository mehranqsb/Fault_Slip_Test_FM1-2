This work presents the quantitative results of a hydro-elasto-plastic model that investigates the interactions between fault reactivation in a clay host rock and the possibility for increased fluid displacement through a low-permeability medium. The work addresses important issues related to potential creation of permeable flow paths for contaminant transport in otherwise low permeability argillaceous rocks. Flow paths can be formed by reactivating faults of different magnitudes through thermal, hydraulic, and mechanical disturbances that occur during the operational or post-closure period. Pore pressure changes caused by thermal pressurization, as demonstrated by Rutqvist et al. in [1], or gas production can potentially reactivate faults with respect to mechanical responses and associated changes in
fault permeability. Therefore, a permeability model base on embedded fractures [2] is applied. The primary aim of this study is to analyze the potential failure in a rock mass with mechanical discontinuities, such as faults, fractures, bedding planes, or other weak planar structures. In general, failure may occur in two ways: a sliding failure along the weaker plane or an intrinsic failure of the rock mass. The rock matrix is expected to exhibit either elastic behavior or brittle failure [3].

Our approach utilizes an elastoplastic Mohr-Coloumb constitutive model, which considers the orientation and presence of weakness (such as weak planes) to accurately characterize shear failure in a rock mass containing mechanical discontinuities such as faults, fractures, bedding planes, or other planar, weak structures. There are two potential causes of failure: the rock mass may fracture spontaneously or it may separate along a weak plane. If the rock material undergoes failure, it should exhibit either elastic or brittle behavior, as indicated by a non-associated Mohr-Coulomb behavior. A sliding failure is indicated when the Coulomb criteria is assessed on a visible plane. Failure can also impact the hydraulic characteristics of the rock mass. The shearing of the weakness plane can generate a conduit for fluid to flow through. Figure 1 depicts the predetermined fault inside the geometrical configuration of host rock. The entire two-dimensional domain is divided into two parts: the host rock, represented by the color blue, and the fault, represented by the color red. The initial fluid pressure is $p_0 = 0.5$ MPa, and the initial stress states are $\sigma_x = 3.3$ MPa, $\sigma_y = 6$ MPa, and $\sigma_z = 7$ MPa. The model assumes that the pressure and stress fields are uniform, and neglects the influence of gravity. This benchmark example involves using two different models, called FM1 and FM2, each with a distinctive parameter set, to study the changes in aperture driven by shear and the resulting variations in fracture transmissivity. The FM1 model implies an increase in the hydraulic conducting aperture due to a predetermined fracture known as the initial creation aperture when shear failure occurs. In this model, the occurrence of shear or tensile failure results in the creation of an aperture with a width of $\Delta b_{\text{i}} = 28 \mu$ m, indicating the formation of an open fault surface.

The model domains have side-lengths of $20$ m and both contain a fault dipping $65$ degrees. The mesh configuration, dimensions of the 3D model and the strcuture of the embedded-fracture fault is represented in Fig. 1.

# Description of Material Properties

This section discusses thoroughly how to calculate the mechanical and hydraulic parameters. 
The host rock is considered an elastic medium, whereas the fault is regarded as a weak plane embedded in elastoplastic media. 
In this study, the hydraulic conductivity of rock cracks is examined using the parallel plate model with cubic law. 
In the model, the fracture is assumed to be confined by two smooth, parallel walls separated by an aperture $b$. 
The flow is supposed to occur in a given fracture, with the flow rate directly proportional to the fluid pressure gradient and gravitational force, and transversely proportional to the dynamic viscosity $\mu$. 


## Mecahnical Propeties

The Young's modulus and the Poisson's ratio of the host rock matrix is calculated regarding the constitutive parameters, given in Table 1 by

$$
\tag{1}
\begin{equation}
E_{\text{M}} = \frac{9\, K\, G}{3\,K + G} \quad \text{and} \quad \nu_{\text{M}}  = \frac{3\,K - 2\,\mu}{2(3\,K + \mu)},
\end{equation}
$$

as $E_{\text{M}} = 6.1065$ GPa and $\nu_{\text{M}} = 0.3275$

Normal stiffness is evaluated considering the thickness of the fault and its Young's modulus accroding to the relations derived in [4]:

$$
\tag{2}
\begin{equation}
K_n = \frac{E\,(1- \nu)}{(1+\nu)(1-2\nu)}\frac{1}{t_{\text{el}}}
\end{equation}
$$

where the mean fracture distance is $a=0.021117$ m. Then, the Young's modulus of the fault is calculated as $E := 281.6$ kPa. Considering that the normal and shear stiffness of the fault are both equal, the Poisson's ratio is determined as $\nu = -0.5$.

## Permeability Model

The fundamental concept of the model revealed here is the suitable depiction of individual fractures integrated into a continuous finite element. A finite element made up of a rock matrix and a number of $n$ fractures is seen in Figure 2. The breadth of each fracture, $a$, a characteristic size of the material, and the element size, $s$, determine the number of fractures in an element.

The fluid flow in the fracture is governed by Darcy's law, with a cubic dependency between the flow rate and the hydraulic conducting aperature. The fluid flow per fracture cross-section $q$ can be expressed as

$$
\tag{3}
\begin{equation}
q_i = -\frac{b^2}{12\,\mu} \left(p_{,i} - \rho_{\text{f}} g_i\right).
\end{equation}
$$

Consider first the flow phenomena through a single fracture. Liquid and gas flow will be calculated using Darcy’s law. The most important parameter in this law
is the intrinsic permeability which can be calculated, assuming laminar flow, as

$$
\tag{4}
\begin{equation}
k_{\text{F}} = \frac{b^3}{12\,t_{\text{el}}},
\end{equation}
$$

where $t_{\text{el}}$ indicates the tickness of the falut-embedded element that scales the permeability to correctly generate the cubic flow rule.

The transmisibility $T$ is the determined by

$$
\tag{5}
\begin{equation}
T_{\text{F}} = k\,s \quad \text{with} \quad k = k_{\text{M}} + \frac{b}{a} k_{\text{F}}.
\end{equation}
$$

The total permeability $k$ of the embedded fracture zone consists of the intrinsic matrix permeability $k_\text{M}$, the ratio of $b$ and the mean fracture distance $a$ as well as the fracture permeability $k_\text{F}$. $k_\text{M}$ is assumed to be nearly impermeable for the current benchmark example. In additon, there applies $s=a$ due to the existance of only one fracture in the benchmark. The expression of $T_{\text{F}} $ simplifies therefore to

$$
\tag{6}
\begin{equation}
T_{\text{F}} \approx \frac{b^2}{12}\,b.
\end{equation}
$$

When a set of $n$ fractures is included in a ﬁnite element, the equivalent intrinsic permeability of the element can be calculated as

$$
\tag{7}
\begin{equation}
k = k_{\text{M}} \boldsymbol{M} + \left( \frac{s-nb}{s}k_{\text{M}} + \frac{nb}{s}k_{\text{F}} \right) (\boldsymbol{I} - \boldsymbol{M} )
\quad \text{with} \quad n = \frac{s}{a}.
\end{equation}
$$

By replacing $k_{\text{F}}$ and $n$ in (7), we obtain the permeability as

$$
\tag{8}
\begin{equation}
k = k_{\text{M}} \boldsymbol{I} + \frac{b}{a} \left(\frac{b^2}{12} - k_{\text{M}} \right) (\boldsymbol{I} - \boldsymbol{M}),
\end{equation}
$$

where the hydraulic conducting aperature (opening) is addititively decoupled into the initial aperature $b_{\text{i}}$, 
an induced fracture creation aperature change, $\Delta b_{\text{c}}$, an induced elastic aperature change, $\Delta b_{\text{e}}$,
an induced shear dilation aperature change, $\Delta b_{\text{s}}$ as

$$
\tag{9}
\begin{equation}
b = b_i + \Delta b_{\text{e}} + \Delta b_{\text{ic}} .
\end{equation}
$$

In the current study, two different fault models, simply called FM1 and FM2, with different parameter sets are applied for dealing with shear-induced hydraulic aperture changes, and associated changes in fracture transmissivity. The models explain the different repsonses with respect to the evolution of the pressure along the fault path. The FM1 model is applied by introducing an initial creation aperture of $\Delta b_{\text{ic}} = 28 \mu$ m that comes from the modeling and evaluation of similar field experiments at the Tournemire site [5]. This parameter indicates a pre-set facture in the FM1 model, which is added as soon as the plastic deformation occures.

The induced elastic aperature change, $\Delta b_{\text{e}}$ is calculated by

$$
\tag{10}
\begin{equation}
\Delta b_{\text{e}} = a \, \langle \varepsilon_{\text{n}} - \varepsilon_{0} \rangle,
\end{equation}
$$

where $\varepsilon_{0}$ indicates the threshold strain, and $\langle \varepsilon_{\text{n}} - \varepsilon_{0} \rangle$ states that the corresponding aperature change occurs only when $\varepsilon_{\text{n}} \gt \varepsilon_{0} $. The thredshold strain $\varepsilon_{0}$ is determined through 

$$
\tag{11}
\begin{equation}
b(t=0)=b_\text{i} + a \, \langle \varepsilon_\text{n}-\varepsilon_\text{0} \rangle,
\end{equation}
$$

while the normal strain $\varepsilon_\text{n}$ and the initial aperture $b_\text{i}$ are initially postulated to be zero. 
Therefore, the thredshould strain can be determined by

$$
\tag{12}
\begin{equation}
\varepsilon_\text{0}= -\frac{b(t=0)}{a}.
\end{equation}
$$

In the current work it is calculated as $\varepsilon_\text{0}:= -\cfrac{10^{-5}}{0.021117} =  -4.7355 \cdot 10^{-4}$.

# Multi-Mechanism Plasticity: Mohr-Coulomb with Coulomb Weakness Planes

This model deals with the existence of a weak orientation (known as a weak plane) within a Mohr-Coulomb model. On this plane, failure criteria are applied, forming a combined Mohr-Coulomb envelope with a tension cutoff. The position of a stress point on this envelope is determined using a non-associated flow rule for shear failure and an associated rule for tension failure. The weak-plane orientation is given by the Cartesian components of a unit normal to the plane in the global $x$-, $y$-, $z$-axes. A local system of reference axes is defined with $e_0$, $e_1$, and $e_3$. In the current formulation, the normal $e_0$ and the tangential vector $e_1$ are calculated with respect to fault-embedded medium dipping $65^\circ$. Therefore, in the local coordinate syste, the basis vectors are $e_0 = (-0.906 0.423)$, and $e_1 = (-0.423, -0.906)$. The Mohr-Coulomb yield and potential functions,

$$
\tag{13}
\begin{equation}
    F^{\mathrm{Matrix}} = \frac{1}{3} \sin \phi + \sqrt{J_2 K^{2}(\theta) + a^2 \sin^2 \phi} - c \cos \phi
\end{equation}
$$

and

$$
\tag{14}
\begin{equation}
    G^{\mathrm{Matrix}}_F = \frac{1}{3} \sin \psi + \sqrt{J_2 K^{2}(\theta) + a_G^2 \sin^2 \psi} - c \cos \psi
\end{equation}
$$

are employed to describe the non-associated behaviour of the matrix, and

$$
\tag{15}
\begin{equation}
    F^{\mathrm{wp}} = \sigma^{\mathrm{wp}}_n \tan \phi^{\mathrm{wp}} - c^{\mathrm{wp}} + \tau^{\mathrm{wp}}
\end{equation}
$$

and

$$
\tag{16}
\begin{equation}
    G^{\mathrm{wp}}_F = \sigma^{\mathrm{wp}}_n \tan \phi^{\mathrm{wp}} - c^{\mathrm{wp}} + \tau^{\mathrm{wp}}
\end{equation}
$$

are used to extend a multi-surface plasticity envolving weakness plane with normal $n^{\mathrm{wp}}$, regarding non-associated Coulomb behaviour. In above equations, the normal and the shear stresses on the fault path are calculated by

$$
\begin{equation}
\tag{17}
\sigma^{\mathrm{wp}}_{\mathrm{n}} = \boldsymbol{\sigma} : \boldsymbol{n} \otimes \boldsymbol{n}
\end{equation}
$$

and

$$
\tag{18}
\begin{equation}
\tau^{\mathrm{wp}}_{\mathrm{n}} = \sqrt{\sigma^{\mathrm{T}}\sigma:\boldsymbol{n} \otimes \boldsymbol{n}- (\boldsymbol{\sigma} : \boldsymbol{n} \otimes \boldsymbol{n})^2}
\end{equation}
$$

## Calculation of the normal and shear displacment along the fault dip

```python
def displacement_ns(file, theta, pt_left, pt_right):
    u_nd_left = file["displacement"][pt_left].T[0] * math.cos(math.radians(theta)) + file[
        "displacement"
    ][pt_left].T[1] * math.sin(math.radians(theta))
    u_sd_left = -file["displacement"][pt_left].T[0] * math.sin(math.radians(theta)) + file[
        "displacement"
    ][pt_left].T[1] * math.cos(math.radians(theta))
    ##
    u_nd_right = file["displacement"][pt_right].T[0] * math.cos(math.radians(theta)) + file[
        "displacement"
    ][pt_right].T[1] * math.sin(math.radians(theta))
    u_sd_right = -file["displacement"][pt_right].T[0] * math.sin(math.radians(theta)) + file[
        "displacement"
    ][pt_right].T[1] * math.cos(math.radians(theta))
    return abs(u_nd_left) + abs(u_nd_right), abs(u_sd_left) + abs(u_sd_right)
```

## Calculation of the shear and normal stresses on the fault dip

```python
def stress_ns(file, pt, theta):
    nVector = np.array(
        [math.cos(math.radians(theta)), math.sin(math.radians(theta)), 0.0]
    )
    # Compute the dyadic product
    dyadic_product_of_the_normal_vector = np.outer(nVector, nVector)
    ndyn_wp = dyadic_product_of_the_normal_vector
    stress_ten = np.zeros((3, 3))
    sigma_wp = np.array([0.0])
    tau_wp = np.array([0.0])
    stress = file[pt]
    print(np.array(stress)[0][0], len(stress_ten))
    for i in range(len(stress)):
        stress_ten[0, 0] = np.array(stress)[i][0]
        stress_ten[1, 1] = np.array(stress)[i][1]
        stress_ten[2, 2] = np.array(stress)[i][2]
        stress_ten[0, 1] = np.array(stress)[i][3]
        stress_ten[1, 0] = np.array(stress)[i][3]
        stress_ten[1, 2] = 0.0
        stress_ten[2, 1] = 0.0
        stress_ten[0, 2] = 0.0
        stress_ten[2, 0] = 0.0
        s_wp = np.tensordot(stress_ten, ndyn_wp)
        t_wp = np.sqrt(
            np.tensordot((np.dot(stress_ten, stress_ten.transpose())), ndyn_wp)
            - s_wp * s_wp
        )
        sigma_wp = np.append(sigma_wp, s_wp)
        tau_wp = np.append(tau_wp, t_wp)
    return sigma_wp, tau_wp
```
